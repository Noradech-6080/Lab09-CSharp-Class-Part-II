## คำถาม

1. Static Method คืออะไร


คือ Method ที่สามารถเรียกใช้ผ่านชื่อ class ได้เลยไม่จำเป็นต้องสร้าง object ของclassมาก่อนการใช้

2. ประกาศ static method ในภาษา C# อย่างไร


ประกาศโดยใช้ Keyword static

3. เรียกใช้ static method อย่างไร


เรียกใช้ผ่านชื่อ class โดยตรง แล้วพิมพ์ชื่อ Method ตาม

4. static method สามารถเข้าถึง instance members (fields, methods) ได้หรือไม่


 ถ้าเข้าถึงโดยตรงจะทำไม่ได้ จะต้องทำการสร้าง object ก่อน แล้วค่อยเข้าถึง instance members ผ่านตัว object instance

5. static method เหมาะกับการใช้งานแบบไหน


 เหมาะกับการใช้งานในโปรแกรมที่ไม่ต้องเข้าถึงข้อมูลสร้าง object หรือ instance

6. ข้อดีของการใช้ static method คืออะไรบ้าง


เรียกใช้งานได้ทันทีไม่ต้องสร้าง object ใช้หน่วยความจำน้อยทำงานเร็วไม่ต้องสร้าง instance

7. ข้อเสียของการใช้ static method คืออะไรบ้าง


ไม่สามารถใช้งานร่วมกับ instance member ของ class ได้

8. จงยกตัวอย่าง static method ที่ใช้บ่อยใน Library มาตรฐานของภาษา C# พร้อมทั้งยกตัวอย่างการใช้งาน

ตัวอย่างการใช้ Array.Sort(), Array.Reverse() และ Array.IndexOf() ใน C#

````
using System;

class Program
{
    static void Main()
    {
        int[] numbers = { 6, 6, 0, 3, 0, 2, 7, 0 };

        // เรียงลำดับจากน้อยไปมาก
        Array.Sort(numbers);
        Console.WriteLine($"Array เรียงลำดับ: {String.Join(", ", numbers)}");

        // กลับด้านอาร์เรย์
        Array.Reverse(numbers);
        Console.WriteLine($"Array กลับด้าน: {String.Join(", ", numbers)}");

        // ค้นหาตำแหน่งของเลข 7
        int index = Array.IndexOf(numbers, 7);
        Console.WriteLine($"ตำแหน่งของเลข 7: {index}");

        // ค้นหาตำแหน่งของเลข 0 ตัวแรก
        int indexZero = Array.IndexOf(numbers, 0);
        Console.WriteLine($"ตำแหน่งของเลข 0 ตัวแรก: {indexZero}");
    }
}
````
